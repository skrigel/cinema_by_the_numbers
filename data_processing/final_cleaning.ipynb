```python
import pandas as pd
import json
import os
```

# Merging Data from TMBD API and Original SOVAI Data


```python
CLEAN_DATA_PATH = "data/cleaned"
RAW_DATA_PATH = "raw"
```

```python
merged_df = pd.read_csv(f'{CLEAN_DATA_PATH}/merged.csv')
merged_df.head()
```

```python
tmdb_movie_metadata = pd.read_csv(f"{RAW_DATA_PATH}/tmdb_movie_metadata.csv")
```

```python
tmdb_movie_metadata['tmdb_id'] = tmdb_movie_metadata['id']
tmdb_movie_metadata = tmdb_movie_metadata.drop(columns=['id'])
```

```python
print(tmdb_movie_metadata.shape)

tmdb_movie_metadata.head()
```

```python
merged_df['tmdb_id'] = merged_df['id']
merged_df = merged_df.drop(columns=['id'])

```

```python
print(merged_df.shape)
merged_df.head()
```

## 1. Merge on TMDB

```python
final_df = merged_df.merge(
    tmdb_movie_metadata,
    on="tmdb_id",
    how="left",  # keeps all rows from merged_df, adds matching metadata where available
    suffixes=("_sov", "_tmdb")  # avoids column name conflicts
)

print(final_df.shape)
final_df.drop(columns=['Unnamed: 0'], inplace=True)


print("Shape: ", final_df.shape)
print(final_df.columns)
```

```python
final_df.to_csv(f"{CLEAN_DATA_PATH}/final_df.csv")
```

# 2. Filtering Data


1. Filter to english language

```python
filtered = final_df[final_df['original_language'] == 'en']
filtered.shape
```

2. Make sure have usable gross info

```python
filtered = filtered[
    (filtered['budget'] > 0) &
    (filtered['gross'] > 0)
]
filtered.shape
```

3. Create new feature (gross per theater)

```python
filtered['gross_per_theater'] = filtered['gross'] / filtered['theaters']
```

4. Ensure inputs are present

```python
filtered = filtered[filtered['genre_ids'
].isna()==False]
print(filtered.shape)
```

```python
filtered['release_date'] = pd.to_datetime(filtered['release_date_sov'])
filtered['release_month'] = filtered['release_date'].dt.month
filtered['release_weekday'] = filtered['release_date'].dt.dayofweek
```
