# Merging Data from TMBD API and Original SOVAI Data


```python
CLEAN_DATA_PATH = "../data/cleaned"
RAW_DATA_PATH = "../data/raw"
```

```python
merged_df = pd.read_csv(f'{CLEAN_DATA_PATH}/merged_df.csv')
merged_df.head()
```

```python
merged_df.rename(columns={"id": "tmdb_id"}, inplace=True)
```

```python
tmdb_movie_metadata = pd.read_csv(f"{RAW_DATA_PATH}/tmdb_movie_metadata.csv")
```

```python
tmdb_movie_metadata['tmdb_id'] = tmdb_movie_metadata['id']
tmdb_movie_metadata = tmdb_movie_metadata.drop(columns=['id'])
```

```python
print(tmdb_movie_metadata.shape)

tmdb_movie_metadata.head()
```

```python
print(merged_df.shape)
merged_df.head()
```

## 1. Merge on TMDB

```python
final_df = merged_df.merge(
    tmdb_movie_metadata,
    on="tmdb_id",
    how="left",  # keeps all rows from merged_df, adds matching metadata where available
    suffixes=("_sov", "_tmdb")  # avoids column name conflicts
)

print(final_df.shape)

print("Shape: ", final_df.shape)
print(final_df.columns)
```

```python
final_df.rename(columns={"title_sov": "title", "release_date_sov": "release_date", "popularity_sov": "popularity","original_title_sov": "original_title" }, inplace=True)
final_df = final_df.drop(columns=['release_date_tmdb', 'title_tmdb', 'popularity_tmdb', 'original_title_tmdb']) 
```

```python
to_remove = ['distributor address', 'distributorwebsite', 'homepage', 'tagline', 'original_title', 'production_company_countries', 'production_country_codes', 'production_country_names']
final_df.drop(columns=to_remove, inplace=True)
```

```python
final_df
```

```python
final_df.drop_duplicates(subset="title", inplace=True)
```

```python
final_df.reset_index(drop=True, inplace=True)
```

```python
final_df.date = pd.to_datetime(final_df.date)
```

```python
final_df.to_csv(f"{CLEAN_DATA_PATH}/merged_with_metadata.csv")
```

```python
final_df=final_df[final_df.date > "1990-01-01"]
```

```python
final_df
```

# 2. Filtering Data


1. Filter to english language

```python
filtered = final_df[final_df['original_language'] == 'en']
filtered.shape
```

2. Make sure have usable gross, runtime, and popularity info

```python
filtered = filtered[
    (filtered['total_gross'] > 0) &
    (filtered['average_gross'] > 0) &
    (filtered['popularity'] > 0) &
    (filtered['runtime'] > 0)
]
print(filtered.shape)
```

3. Create new feature (gross per theater)

```python
filtered['gross_per_theater'] = filtered['total_gross'] / filtered['theaters']
```

4. Ensure inputs are present

```python
filtered = filtered[filtered['genre_ids'
].isna()==False]
print(filtered.shape)
```

```python
filtered.to_csv('../data/cleaned/final_df.csv', index=False)
```
